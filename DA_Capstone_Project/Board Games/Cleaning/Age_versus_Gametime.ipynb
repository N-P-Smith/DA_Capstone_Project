```python
df1_1 = df1.groupby("min_age")["min_time"].min().round().astype(int)

df1_1
```

```python
df1_2 = df1.groupby("min_age")["min_time"].mean().round().astype(int)

df1_2
```

```python
df1_3 = df1.groupby("min_age")["max_time"].mean().round().astype(int)

df1_3
```

```python
df1_4 = df1.groupby("min_age")["max_time"].max().round()

df1_4
```

```python
df1_5 = pd.merge(df1_1, df1_2, how = "outer", on = "min_age")
df1_6 = pd.merge(df1_5, df1_3, how = "outer", on = "min_age")
df2 = pd.merge(df1_6, df1_4, how = "outer", on = "min_age")

df2
```

```python
df2.columns = ["Minimum Play Time", "Average Minimum Time", "Average Maximum Time", "Maximum Play Time"]

df2
```

```python
df2.to_csv("Age_versus_Gametime.csv")
```

```python
plt.figure(figsize=(12,5)) 

sns.scatterplot(data = df2)
```

```python
df3 = df[["year", "min_age", "min_time", "max_time"]].dropna(axis = 0).astype(int)
df3 = df3[df3["year"] >= 1900]
df3 = df3[df3["min_age"] <= 21]
df3 = df3[df3["max_time"] <= 360]

df3
```

```python
df3_1 = df3.groupby("min_age")["min_time"].min().round().astype(int)
df3_2 = df3.groupby("min_age")["min_time"].mean().round().astype(int)
df3_3 = df3.groupby("min_age")["max_time"].mean().round().astype(int)
df3_4 = df3.groupby("min_age")["max_time"].max().round().astype(int)

df3_5 = pd.merge(df3_1, df3_2, how = "outer", on = "min_age")
df3_6 = pd.merge(df3_5, df3_3, how = "outer", on = "min_age")
df4 = pd.merge(df3_6, df3_4, how = "outer", on = "min_age")

df4.columns = ["Minimum Play Time", "Average Minimum Time", "Average Maximum Time", "Maximum Play Time"]

df4
```

```python
plt.figure(figsize=(12,5)) 

sns.scatterplot(data = df4)
```
