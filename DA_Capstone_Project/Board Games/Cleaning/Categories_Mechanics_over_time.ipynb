```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../Data/details.csv")

df.info()
```

```python
df1 = df[["primary", "yearpublished", "boardgamecategory"]].dropna(axis = 0)
df1 = df1[df1["yearpublished"] > 1900]

df1
```

```python
df1["boardgamecategory"][1:-1].replace("'", "")

df1
```

```python
unique_categories = []

for row in df1["boardgamecategory"]:
    for boardgamecategory in row.split(","):
        if boardgamecategory not in unique_categories:
            unique_categories.append(boardgamecategory)

unique_categories
```

```python
for boardgamecategory in unique_categories:
    df1[boardgamecategory] = 0

df1
```

```python
for index, row in df1.iterrows():
    categories = row["boardgamecategory"].split(",")
    for boardgamecategory in categories:
        df1.at[index, boardgamecategory] = 1

df1
```
