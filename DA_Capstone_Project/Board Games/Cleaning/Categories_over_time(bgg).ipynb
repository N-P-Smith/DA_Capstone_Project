```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../Data/bgg_GameItem.csv")

df.info()
```

```python
df1 = df[["name", "year", "category"]].dropna(axis = 0)
df1 = df1[df1["year"] >= 1900]
df1["year"] = df1["year"].astype(int)


df1
```

```python
df1.info()
```

```python
df1["category"].info()
```

```python
unique_categories = []  # Start with an empty list

# Loop through each row to collect unique categories
for row in df1["category"]:
    for category in row.split(','):
        if category not in unique_categories:  # Ensure uniqueness
            unique_categories.append(category)
```

```python
unique_categories
```

```python
for category in unique_categories:
    df1[category] = 0
```

```python
df1.head()
```

```python
for index, row in df1.iterrows():  #loop through each row
    categories = row['category'].split(',') #splitting the cat
    for category in categories:
        df1.at[index, category] = 1
```

```python
df1.head()
```

```python
df1.to_csv("Categories_over_Time.csv")
```

```python
# Read the categories CSV file
category_names = pd.read_csv("../Data/bgg_Category.csv")

category_names.info()
```

```python
df1_rename = dict(zip(category_names["bgg_id"].astype(str), category_names["name"]))

df1_rename
```

## Categories as a column version.

```python
df1_melt = df1.melt(id_vars = ["name", "year", "category"], value_vars = df1.columns[3:], var_name = "category_name", value_name = "exists")

df1_melt
```

```python
df1_melt.columns = ['name', 'year', 'category_set', 'category', 'exists']

df1_melt
```

```python
df1_melt_cleaned = df1_melt[df1_melt["exists"] == 1].drop(columns = ["exists"]).sort_values(["year", "name"])

df1_melt_cleaned
```

```python
df1_melt_cleaned.to_csv("Categories_over_Time_melt.csv")
```
