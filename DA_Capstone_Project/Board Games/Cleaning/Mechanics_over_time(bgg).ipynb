```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../Data/bgg_GameItem.csv")

df.info()
```

```python
df2 = df[["name", "year", "mechanic"]].dropna(axis = 0)
df2 = df2[df2["year"] >= 1900]
df2["year"] = df2["year"].astype(int)

df2
```

```python
unique_mechanics = []

for row in df2["mechanic"]:
    for mechanic in row.split(","):
        if mechanic not in unique_mechanics:
            unique_mechanics.append(mechanic)

unique_mechanics
```

```python
for mechanic in unique_mechanics:
    df2[mechanic] = 0

df2.head()
```

```python
for index, row in df2.iterrows():
    mechanics = row["mechanic"].split(",")
    for mechanic in mechanics:
        df2.at[index, mechanic] = 1

#for row in df2["mechanic"]:
#      row_mechanics = row.split(",") 
#for mechanic in unique_mechanics: 
#   mechanic_columns[mechanic].append(1 if mechanic in row_mechanics else 0)
   
df2
```

```python
mechanic_names = pd.read_csv("../Data/bgg_Mechanic.csv")

mechanic_names.info()
```

```python
df2_rename = dict(zip(mechanic_names["bgg_id"].astype(str), mechanic_names["name"]))

df2.rename(columns = df2_rename, inplace=True)

df2
```

```python
df2.to_csv("Mechanics_over_Time_labelled.csv")
```

```python
df2_melt = df2.melt(id_vars = ["name", "year", "mechanic"], value_vars = df2.columns[3:], var_name = "mechanic_name", value_name = "exists")

df2_melt
```

```python
df2_melt.columns = ["Name", "Year", "Mechanics Set", "Mechanic", "exists"]

df2_melt
```

```python
df2_melt_cleaned = df2_melt[df2_melt["exists"] == 1].drop(columns = ["exists"]).sort_values(["Year", "Name"])

df2_melt_cleaned
```

```python
df2_melt_cleaned.to_csv("Mechanics_over_Time_melt.csv")
```
